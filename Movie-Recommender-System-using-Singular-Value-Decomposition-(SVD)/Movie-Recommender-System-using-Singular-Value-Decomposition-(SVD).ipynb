#### Create m x u matrix with movies as row and users as column. Normalize the matrix.

```python
rating_mat = np.ndarray(shape = (np.max(rating_df['movie_id'].values), np.max(rating_df['user_id'].values)))
rating_mat[rating_df['movie_id'].values -1, rating_df['user_id'].values -1] = rating_df['rating'].values 
print(rating_mat)
```

```python
rating_mat.shape
```

```python
matrix_normed = normalize(rating_mat, axis=1, norm='l1')
print(matrix_normed)
```

```python
row_sums = np.sum(matrix_normed, axis=1)
print(row_sums)
```

#### Perform SVD to get U, S and V.

```python
A = matrix_normed.T
print(A)
```

```python
U, S, V = np.linalg.svd(A)
U.shape, S.shape, V.shape
```

```python
print(U)
```

```python
print(V)
```

```python
print(S)
```

```python
sigma = np.zeros((6040,3952))
for i in range(3952):
    sigma[i,i] = S[i]
print(sigma)
```

#### Select top 50 components from S.

```python
S50 = sigma[0:50,0:50]
print(S50)
```

```python
U50 = U[:,0:50]
```

```python
V50 =V[0:50,:]
```

```python
AR50 = U50@S50@V50
print(np.round(AR50))
print(AR50.shape)
print(AR50.size)
```

```python
sum_AR50 = np.sum(AR50, axis=0)
print(sum_AR50)
```

#### Get the top 50 eigenvectors using eigenvalues.

```python
evalue, evector = nla.eig(S50)
```

#### Using cosine similarity, find 10 closest movies using the 50 components from SVD.

```python
import numpy as np
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

# Assuming you have a movie-user matrix named 'ratings_matrix'
# Replace this with your actual movie-user matrix

# Perform SVD
svd = TruncatedSVD(n_components=50)
print(svd)
svd_matrix = svd.fit_transform(A)
print(svd_matrix.shape)
# Calculate cosine similarity
cosine_sim = cosine_similarity(svd_matrix, svd_matrix)

# Find 10 closest movies for each movie
closest_movies = {}
for i, movie_row in enumerate(cosine_sim):
    closest_movies[i] = sorted(enumerate(movie_row), key=lambda x: x[1], reverse=True)[1:11]

# Display the results
for movie_id, closest in closest_movies.items():
    print(f"\nClosest movies to Movie {movie_id + 1}:")
    for close_movie_id, similarity in closest:
        print(f"Movie {close_movie_id + 1}, Similarity: {similarity:.4f}")
```
